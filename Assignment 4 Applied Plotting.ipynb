```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data source = http://data.un.org/_Docs/SYB/CSV/SYB62_319_201906_Ratio%20of%20Girls%20to%20Boys%20in%20Education.csv
#second = http://data.un.org/_Docs/SYB/CSV/SYB62_317_201905_Seats%20held%20by%20Women%20in%20Parliament.csv
```

```python
df = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\Ratio of Girls to Boys in Education.csv', skiprows=1)
```

```python
df.head()
```

```python
df =df[['Unnamed: 1','Year', 'Series','Value' ]] 
```

```python
df.head()
```

```python
df.info()
```

```python
df['Series'].unique()
```

```python
designation  = {'Ratio of girls to boys in primary education': 1, 
                'Ratio of girls to boys in secondary education': 2,
                'Ratio of girls to boys in tertiary education' : 3}
```

```python
df['Series'] = df['Series'].map(designation)
```

```python
df = df.iloc[330:,]
```

```python
df.head(20)
```

```python
df['Year'].value_counts()
```

```python
df.columns = ['Country', 'Year', 'School Type', 'Ratio']
```

```python
df['Country'].value_counts()
```

```python
df.set_index(['Year', 'School Type'])
```

```python
df.reset_index()
```

```python
df.head()
```

```python
india = df[df['Country']=='India']
```

```python
india['School Type'] = india['School Type'].astype(str) 
```

```python
reverse  = { '1': 'Primary School', 
             '2': 'Secondary School',
             '3': 'Tertiary (college)'}
india['School Type'] = india['School Type'].map(reverse)
```

```python
india
```

```python
sns.set_style('white')
fig = sns.lmplot(x='Year', y='Ratio',hue='School Type', data=india,fit_reg=False, legend=False);
plt.subplots_adjust(top=0.95);
plt.title("India's Girl-Boy Ratio over the Year 1985-2015 " );
plt.legend(title='Education Level',loc=4) #bbox_to_anchor=(1.4,0.5));
```
