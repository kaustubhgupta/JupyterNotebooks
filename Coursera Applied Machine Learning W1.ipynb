# A simple classification using k-nn

```python
df
```

```python
df = pd.read_table('D:\\Coursera Data Science\\Course 3\\fruit_data_with_colors.txt')
```

```python
X = df[['mass','width','height']]
y = df[['fruit_label']]

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
y_train.shape
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
lookup = dict(zip(df.fruit_label.unique(), df.fruit_name.unique()))
```

```python
pre = knn.predict([[20, 4.5, 2]])
lookup[pre[0]]
```

# sensitivity of k-nn wrt choice of k

```python
k_range = range(1,20)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
plt.xticks([0,5,10,15,20]);
```
